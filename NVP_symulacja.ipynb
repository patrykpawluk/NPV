```python
def calc_max_commission(K, start_date, repayment_date):
    """
    K - kapitał pożyczki
    start_date - pierwszy dzień trwania pożyczki
    repayment_date - dzień płatności ostatniej raty
    
    MPKK = MIN(0.45 * K, (K x 10%) + (K x n/R x 10%))
    Wersja lekko zmodyfikowana pełne lata dadzą n/R = 1 
    pozostaje reszta z dzielenia n/R wyznaczająca liczbę dni 
    i R jako liczbę dni w roku przypadającą w ostatnim roku umowy pożyczki
    Stąd: 
    MPKK = (K x 10%) + (K x (N + n/R) x 10%)
    Gdzie:
    K - kapitał pożyczki
    N - liczba pełnych rat obowiązywania kredytu
    n - liczba dni pozostała do spłaty w niepełnym ostatnim roku życia pożyczki
    R - liczba dni w roku w którym pożyczki ma swój koniec
    """
    from datetime import datetime
    from dateutil.relativedelta import relativedelta
    from math import floor
    
    time_delta = relativedelta(repayment_date, start_date)

    N = time_delta.years
    n = (repayment_date - (start_date + relativedelta(years=time_delta.years))).days
    R = (datetime(repayment_date.year+1, 1, 1) - datetime(repayment_date.year, 1, 1)).days

    MPKK = floor(min(0.45*K, (K * 0.1) + (K * (N + n/R) * 0.1))*100)/100
    
    return MPKK
```

```python
def calculate_npv_full(df, capital, annual_discount_rate):
    monthly_rate = (annual_discount_rate) / 12

    npv = -capital  # initial loan outflow

    for _, row in df.iterrows():
        t = row["installment_number"]
        cf = row["total_installment"]

        npv += cf / (1 + monthly_rate) ** t

    return npv
```

```python
import pandas as pd
from math import floor

def loan_schedule_commission_first(capital, installments, annual_interest_rate, commission):
    monthly_rate = (annual_interest_rate ) / 12

    # Standard annuity payment (ONLY for capital)
    if monthly_rate == 0:
        base_installment = capital / installments
    else:
        base_installment = (capital+commission) * (
            monthly_rate * (1 + monthly_rate) ** installments
        ) / (
            (1 + monthly_rate) ** installments - 1
        )
        

    itter = 0
    repeat = 1
    while repeat == 1:
        itter += 1
        balance = capital
        remaining_commission = commission
        schedule = []
        print(base_installment)
        for i in range(1, installments + 1):
            interest = balance * monthly_rate
    
            payment_left = base_installment
    
            # 1. Pay interest first
            interest_paid = min(payment_left, interest)
            payment_left -= interest_paid
    
            # 2. Pay commission
            commission_paid = min(payment_left, remaining_commission)
            payment_left -= commission_paid
            remaining_commission -= commission_paid
    
            # 3. Remaining goes to principal
            principal_paid = payment_left
    
            # Prevent overpaying principal in last period
            if principal_paid > balance:
                principal_paid = balance
    
            balance -= principal_paid
    
            total_installment = interest_paid + commission_paid + principal_paid
    
            schedule.append({
                "installment_number": i,
                "remaining_balance": round(balance, 2),
                "interest": round(interest_paid, 2),
                "commission": round(commission_paid, 2),
                "principal": round(principal_paid, 2),
                "total_installment": round(total_installment, 2),
                "remaining_commission": round(remaining_commission, 2)
            })

        if floor(total_installment*100) < floor(base_installment*100):
            #print("total_installment < base_installment")
            base_installment = base_installment-((base_installment-total_installment)/(installments-1))
            repeat = 1
        elif floor(balance*100) != 0:
            #print("balance != 0")
            base_installment = base_installment+(balance/(installments))
        else:
            repeat = 0
            print("Last itter: ",itter)

        if itter > 30:
            repeat = 0
            print("Max itter triggered: ",itter)
            break

    return pd.DataFrame(schedule)
```

```python
import pandas as pd
from math import floor

def loan_schedule_commission_first_ballon(
    capital, 
    installments, 
    annual_interest_rate, 
    commission,
    balloon_amount=0
    ):
    monthly_rate = (annual_interest_rate ) / 12
    target_final_balance = balloon_amount

    # Standard annuity payment (ONLY for capital)
    if monthly_rate == 0:
        base_installment = capital / installments
    else:
        base_installment = (capital+commission-target_final_balance) * (
            monthly_rate * (1 + monthly_rate) ** installments
        ) / (
            (1 + monthly_rate) ** installments - 1
        )

    itter = 0
    repeat = 1
    while repeat == 1:
        itter += 1
        balance = capital
        remaining_commission = commission
        schedule = []
        print(base_installment)
        for i in range(1, installments + 1):
            interest = balance * monthly_rate
    
            payment_left = base_installment
    
            # 1. Pay interest first
            interest_paid = min(payment_left, interest)
            payment_left -= interest_paid
    
            # 2. Pay commission
            commission_paid = min(payment_left, remaining_commission)
            payment_left -= commission_paid
            remaining_commission -= commission_paid
    
            # 3. Remaining goes to principal
            principal_paid = payment_left
    
            # Prevent overpaying principal in last period
            if principal_paid > balance:
                principal_paid = balance
    
            balance -= principal_paid
    
            total_installment = interest_paid + commission_paid + principal_paid
    
            schedule.append({
                "installment_number": i,
                "remaining_balance": round(balance, 2),
                "interest": round(interest_paid, 2),
                "commission": round(commission_paid, 2),
                "principal": round(principal_paid, 2),
                "total_installment": round(total_installment, 2),
                "remaining_commission": round(remaining_commission, 2)
            })

        if floor(balance*100) < floor((target_final_balance)*100):
            print("balance < base_installment+target_final_balance")
            base_installment = base_installment-((target_final_balance-balance)/(installments-1))
            repeat = 1
        elif floor(balance*100) != floor(target_final_balance*100):
            print("balance != 0")
            base_installment = base_installment+((balance-target_final_balance)/(installments))
        else:
            repeat = 0
            print("Last itter: ",itter)

        if itter > 30:
            repeat = 0
            print("Max itter triggered: ",itter)
            break

    return pd.DataFrame(schedule)
```

```python
import datetime as dt
from datetime import datetime
from dateutil.relativedelta import relativedelta

capital_init = 10000
installments=36
start_date = datetime.strptime('2026-01-01', "%Y-%m-%d")
repayment_date = start_date + relativedelta(months=36) 
commision_init = calc_max_commission(capital_init, start_date, repayment_date)

print('Commission',str(commision_init))

df = loan_schedule_commission_first_ballon(
    capital=capital_init,
    installments=installments,
    annual_interest_rate=0.145,
    commission=commision_init,
    balloon_amount=commision_init/4
)

print('Suma prowizji z harmonogramu',str(df.commission.sum()))
print(df)
NPV = calculate_npv_full(df, capital_init, 0.3)
print('NPV kwota',str(NPV))
print('NPV %',str(100*round(NPV/capital_init,7)))
print(df.principal.sum())
```
